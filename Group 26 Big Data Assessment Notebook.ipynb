Change the values in the `CY_Qtr` variable to be just the quarter (e.g. 2025Q1 become just Q1) becuase there is a date column already

```python
pbj_psdf["CY_Qtr"] = pbj_psdf["CY_Qtr"].apply(lambda x: x[len(x)-2 : ]) #change the values in the CY_Qtr variable to be just the quarter
pbj_psdf["CY_Qtr"].head()
```

Convert the `WorkDate` Column into **Date** datatype.

```python
def toDate(val):
    #first split the string into "year", "month", "day" 
    year = str(val)[:4]
    month = str(val)[4:6]
    day = str(val)[6:8]
    #then convert the string to a date
    return dt.datetime(int(year), int(month), int(day))

#convert the WorkDate column to a date
pbj_psdf["WorkDate"] = pbj_psdf["WorkDate"].apply(toDate)
pbj_psdf["WorkDate"].head()
```

**Convert the dataframe back to PySpark Dataframe**

```python
pbj_psdf = pbj_psdf.reset_index(drop=True) #reset the index and drop the duplicate old index as Spark will not be able to carry it over
pbj_clean_df = pbj_psdf.to_spark() #convert to PySpark DataFrame
```

```python
display(pbj_clean_df.limit(5)) #show first 5
```

Renaming some columns for better representation

```python
col_mapping = {
    "PROVNUM": "PROVIDER_ID",
    "PROVNAME": "PROVIDER",
    "STATE": "STATE_CODE",
    "CY_Qtr": "QTR",
    "MDScensus": "RESIDENTS_COUNT"
}

pbj_clean_df = pbj_clean_df.select([pbj_clean_df[c].alias(col_mapping.get(c, c)) for c in pbj_clean_df.columns])
```

Checking the final cleaned dataset

```python
display(pbj_clean_df.limit(5))
```

Converting the Dataframe from Pandas to PySpark's own has led to the `WorkDate` Column's format to changed to Timestamp instead of date. This will be corrected by formating the column as date only.

```python
pbj_clean_df = pbj_clean_df.withColumn("WorkDate", to_date(col("WorkDate"), "yyyy-MM-dd"))#correct the formating of the 'WorkDate' column to Date only not Timestamp
display(pbj_clean_df.limit(10))
```

### Section B: Analysing the Big Dataset
Objectives of the Analysis:
* Understand the average direct care staffing hours to patient ratio for each State
* Analyse the monthly trend of staffing hours in the top 5 states
* Analysing the daily trend of staffing hours in the U.S

**1) Understand the average direct care staffing hours to patient ratio for each State**

```python
#aggregate the total hours worked direct-care job-types only since that is the focus of the analysis.
analyse_df = analyse_df.withColumn("Total_staffing_Hours", col("Hrs_CNA") + col("Hrs_LPN") + col("Hrs_RN"))
```

```python
print("Averaqge Total Daily Staffing Hours per State")

states_avg_staff_hrs_df = analyse_df.groupBy("STATE_CODE").agg(avg("Total_staffing_Hours")).orderBy(desc("avg(Total_staffing_Hours)")).withColumnRenamed("avg(Total_staffing_Hours)", "Mean_Staffing_Hrs")

states_avg_staff_hrs_df.display()
```

```python
states_avg_residents_df = analyse_df.groupBy("STATE_CODE").agg(avg("RESIDENTS_COUNT")).orderBy(desc("avg(RESIDENTS_COUNT)")).withColumnRenamed("avg(RESIDENTS_COUNT)", "Mean_Daily_Residents")

states_avg_residents_df.display()
```

```python
#join the two datasets to get the average daily staffing hours per state and the average daily residents per state
states_staff_hrs_resident_df = states_avg_staff_hrs_df.join(states_avg_residents_df, "STATE_CODE")

#calculate the ratio of staffing hours to patient
states_staff_hrs_resident_df = states_staff_hrs_resident_df.withColumn("Staffing_to_Patient_Ratio", col("Mean_Staffing_Hrs") / col("Mean_Daily_Residents"))

states_staff_hrs_resident_df.display()
```

```python
print("Staffing-to-Resident Ratio per State in 2025 (Hours per Patient)")
states_staff_hrs_resident_df.display()
```

**2) Analysing the monthly trend of the top 5 states with the best staff-to-patient ratio**

```python
top_5_states = states_staff_hrs_resident_df.orderBy(desc("Staffing_to_Patient_Ratio")).limit(5)
top_5_states.display()
```

```python
#Aggregate the daily trend per states for the top 5 states
top_5_states_df = analyse_df.join(top_5_states, "STATE_CODE")
top_5_states_df = top_5_states_df.groupBy("STATE_CODE", "WorkDate").agg(avg("Total_staffing_Hours")).orderBy("WorkDate").withColumnRenamed("avg(Total_staffing_Hours)", "Mean_Staffing_Hrs")

print("Monthly Trend of Staffing Hours in the Top 5 States")
top_5_states_df.display()
```

**3) Analysing the daily trend of staffing hours in the U.S**

```python
#get the name of the day in the week and aggregate the daily trend across the U.S
analyse_states_daily_df = analyse_df.withColumn("Day_of_Week", date_format(col("WorkDate"), "E"))
analyse_states_daily_df = analyse_states_daily_df.groupBy("STATE_CODE", "Day_of_Week").agg(avg("Total_staffing_Hours")).orderBy("Day_of_Week").withColumnRenamed("avg(Total_staffing_Hours)", "Mean_Staffing_Hrs")

print("Daily Average Trend of Staffing Hours in the U.S")
analyse_states_daily_df.display()
```
